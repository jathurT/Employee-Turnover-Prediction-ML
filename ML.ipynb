# Introduction

Employee turnover is the number of employees who leave a company over a specific period of time. Employee turnover prediction is a critical task for organizations to understand the factors contributing to employee attrition and take proactive measures to improve retention. In this project, we present an analysis of employee turnover using a dataset that captures key factors associated with employees' tenure at a company. The dataset includes various attributes that may influence an employee's decision to leave the company, such as their satisfaction level, performance evaluations, work conditions, and compensation.

Our dataset used for this analysis contains the following features:

1. **Satisfaction Level**: A numerical value representing the employee's satisfaction with their job. Lower satisfaction levels have been found to correlate with higher turnover rates.

2. **Last Evaluation:** A numerical value reflecting the most recent performance evaluation score. A lower evaluation score may indicate dissatisfaction or underperformance, potentially leading to attrition.

3. **Number of Projects:** The number of projects an employee is involved in. A higher workload could lead to stress or burnout, which could influence turnover.

4. **Average Monthly Hours:** The average number of hours worked per month. Excessive working hours can contribute to employee burnout and job dissatisfaction, increasing the likelihood of turnover.

5. **Time Spent at Company:** The number of years the employee has been with the company. Employees with longer tenures may have a stronger connection to the organization, while those with shorter tenures might be more prone to leaving.

6. **Work Accident:** A binary indicator of whether the employee had a work-related accident. Such incidents can lead to dissatisfaction or resignation due to perceived safety concerns.

7. **Promotion in the Last 5 Years:** A binary feature indicating whether the employee received a promotion in the last five years. Employees who have not received promotions might feel underappreciated or stagnant, increasing the likelihood of leaving the company.

8. **Department:** The department in which the employee works (e.g., sales, technical, etc.). Certain departments may experience higher turnover rates based on organizational dynamics or job roles.

9. **Salary:** The employee's salary level, categorized as low, medium, or high. Salary disparities can be a significant factor in employee dissatisfaction and turnover.

The target variable in this dataset is "left", which indicates whether the employee has left the company (1) or not (0). The goal of this machine learning project is to use the available features to predict whether an employee is likely to leave the company.

Through this analysis, we aim to uncover patterns and trends that contribute to employee attrition, which can inform strategies for improving employee retention, job satisfaction, and overall organizational performance.

# Literature Survey

Employee turnover is a widely studied phenomenon in human resources (HR) management, as it can have significant financial and operational implications for organizations. A comprehensive body of literature has emerged that investigates the various factors influencing employee attrition. Understanding these factors is crucial for predicting turnover and devising strategies to mitigate it. Below, we review key studies that explore the determinants of employee turnover, focusing on aspects relevant to the dataset used in this analysis.

1. **Job Satisfaction and Employee Turnover**
A well-established factor influencing employee turnover is job satisfaction. According to Mobley (1977), dissatisfaction with one’s job is a primary driver of voluntary turnover. Employees who report low job satisfaction tend to be less committed to their organization, which increases the likelihood of them leaving. In the dataset, satisfaction_level serves as a key indicator of employee contentment. Previous studies have shown a negative correlation between satisfaction and turnover, where lower satisfaction levels correlate with a higher probability of employees leaving the organization (Tett & Meyer, 1993).

2. **Performance Evaluation and Turnover**
Employee performance evaluations play an important role in shaping both employee satisfaction and turnover decisions. Judge & Bono (2001) highlight that poor performance evaluations may lead to feelings of underappreciation and dissatisfaction, potentially prompting employees to seek alternative employment. In the dataset, last_evaluation reflects recent performance assessments, and research suggests that employees with lower evaluation scores are more likely to leave an organization, especially if they perceive no opportunity for growth or promotion (Riley, 2001).

3. **Workload and Burnout**
Excessive workload is another critical factor influencing turnover. The number of projects an employee is involved in and the average monthly hours worked are indicators of workload. Studies by Maslach & Leiter (1997) and Freudenberger (1974) indicate that high workloads and long working hours are linked to job burnout, stress, and eventually turnover. Employees who are overburdened with tasks may experience fatigue, decreased productivity, and lower job satisfaction, increasing their likelihood of leaving the company.

4. **Tenure and Employee Retention**
The time spent at the company is an important predictor of turnover. Employees with longer tenures are often more committed and integrated into the organization, while those with shorter tenures may feel disconnected or face challenges adjusting to the organizational culture. Hom & Griffeth (1995) found that tenure negatively correlates with turnover, where employees with more years of service tend to stay longer. However, Allen et al. (2010) suggest that after a certain point, tenure can have a diminishing effect on retention, particularly if employees perceive limited career advancement opportunities.

5. **Promotion and Career Development**
Promotion opportunities are strongly linked to employee retention. Mobley (1977) found that employees who were not promoted within a reasonable timeframe were more likely to leave, as they perceived a lack of career progression. In the dataset, promotion_last_5years serves as an indicator of whether an employee has received a promotion, which is a critical factor in retention decisions. Employees who feel stagnant in their careers or have not been rewarded for their contributions may seek opportunities elsewhere.

6. **Workplace Safety and Accidents**
The occurrence of a work accident has been shown to influence employee turnover, particularly in hazardous work environments. Kurtz (2007) suggests that employees who experience workplace accidents or witness such incidents may feel insecure about their safety, leading to dissatisfaction and increased turnover. Employees who suffer from work-related injuries or who perceive their work environment as unsafe may be more inclined to leave the organization.

7. **Department and Turnover**
Turnover rates may vary significantly between different departments within an organization. For instance, departments with more stressful or demanding roles, such as sales or customer service, may have higher turnover rates compared to departments with less demanding work. Kacmar et al. (2009) highlight that organizational factors such as departmental work culture, leadership, and job role impact employee satisfaction and turnover rates. The Department feature in the dataset reflects these organizational dynamics, offering insight into the variance in turnover across different business units.

8. **Salary and Employee Retention**
Lastly, salary is a fundamental factor in employee retention. Employees who feel they are underpaid relative to their peers or the market are more likely to seek employment elsewhere. Holtom et al. (2008) found that salary dissatisfaction is one of the strongest predictors of voluntary turnover. The dataset categorizes employees into three salary levels: low, medium, and high. Research suggests that employees in lower salary brackets are generally more prone to turnover, especially if they perceive better compensation opportunities in other organizations.

Summary of Literature
The literature on employee turnover emphasizes several key predictors: job satisfaction, performance evaluations, workload, tenure, promotion opportunities, work accidents, departmental factors, and salary. The dataset used in this study captures many of these dimensions, providing a rich basis for predicting employee turnover. By analyzing these factors, organizations can better understand the underlying causes of turnover and take steps to retain valuable employees, ultimately improving overall organizational performance and reducing the costs associated with employee attrition.

Chi-square Test for categorical features in a dataset.

---




```python
from scipy.stats import chi2_contingency

def chi_square_test_with_target(df, target='left'):
    df_temp = df.copy()
    df_temp[target] = df_temp[target].astype('object')
    categorical_columns = df_temp.select_dtypes(include=['object']).columns.tolist()

    if target not in categorical_columns:
        raise ValueError(f"The target '{target}' is not a categorical column. Make sure it exists and is categorical.")

    p_values = {}
    for col in categorical_columns:
        if col != target:
            contingency_table = pd.crosstab(df_temp[col], df_temp[target])
            chi2, p, dof, expected = chi2_contingency(contingency_table)
            p_values[col] = p

    return p_values
p_values_dict = chi_square_test_with_target(df, target='left')
p_values_df = pd.DataFrame.from_dict(p_values_dict, orient='index', columns=['p_value'])
plt.figure(figsize=(5, len(p_values_df)/2))
sns.heatmap(p_values_df, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, square=False)
plt.title("Chi-Square Test P-Values with Target 'left'")
plt.yticks(rotation=0)
plt.show()
```

It indicates that there is no siginificant connection between categorical variables and target variable

Following the log transformation of the 'Age' variable, we will now assess the data for the presence of any outliers. This step will help ensure that the transformed distribution is suitable for further analysis and modeling

```python
def calculate_fences(series, multiplier=1.5):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - (multiplier * IQR)
    upper_fence = Q3 + (multiplier * IQR)
    return lower_fence, upper_fence

stag_lower_fence, stag_upper_fence = calculate_fences(X_train['time_spend_company'], multiplier=1.5)
print(f"time_spend_company outliers are values < {stag_lower_fence} or > {stag_upper_fence}")
```

Here, eventhough we applied log transformation, we still have outliers in the 'time_spend_company' column. We will cap these outliers by setting them to the nearest acceptable value.

```python
def plot_boxplots(df_train, df_test, columns, titles):
    num_cols = len(columns)
    plt.figure(figsize=(12, 6))

    for i, col in enumerate(columns, 1):
        plt.subplot(1, num_cols * 2, i)
        sns.boxplot(y=df_train[col], color='skyblue')
        plt.title(f'{titles[i - 1]} (Train)')

        plt.subplot(1, num_cols * 2, i + num_cols)
        sns.boxplot(y=df_test[col], color='lightgreen')
        plt.title(f'{titles[i - 1]} (Test)')

    plt.tight_layout()
    plt.show()

columns = ['time_spend_company']
titles_before = ['Stag Before Capping']
plot_boxplots(X_train, X_test, columns, titles_before)

def cap_outliers(df, column, lower, upper):
    return df[column].clip(lower=lower, upper=upper)

X_train['time_spend_company'] = cap_outliers(X_train, 'time_spend_company', stag_lower_fence, stag_upper_fence)
X_test['time_spend_company'] = cap_outliers(X_test, 'time_spend_company', stag_lower_fence, stag_upper_fence)

titles_after = ['time_spend_company After Capping']
plot_boxplots(X_train, X_test, columns, titles_after)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, make_scorer

f1_scorer = make_scorer(f1_score, average='macro')
rf = RandomForestClassifier(random_state=0)
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
}

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    verbose=2,
    scoring=f1_scorer
)
grid_search.fit(X_train, y_train)
print(f"Best Hyperparameters: {grid_search.best_params_}")
print(f"Best Cross-validation F1-Score: {grid_search.best_score_}")
best_rf_model = grid_search.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
test_f1 = f1_score(y_test, y_pred_rf, average='macro')
```

The best cross-validation score obtained during this process was 0.985, indicating that the model performed with 98.5% accuracy on the validation folds during training.

## K-Nearest Neighbor(K-NN) Algorithm


```python
from sklearn.metrics import f1_score, make_scorer
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'p': [1, 2]
}
f1_scorer = make_scorer(f1_score, average='macro')
grid_search = GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    verbose=2,
    scoring=f1_scorer
)

grid_search.fit(X_train, y_train)
print(f"Best Hyperparameters: {grid_search.best_params_}")
print(f"Best Cross-validation F1-Score: {grid_search.best_score_:.4f}")
best_knn_model = grid_search.best_estimator_
y_pred_knn = best_knn_model.predict(X_test)
```
